```python
# checking for nulls
df.isnull().sum()
```

```python
# visualize the nulls
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cmap="BuPu")
plt.show()
```

```python
# impute the nulls if any
df["Age"]=df["Age"].fillna(mn)  # mean
df["Age"]=df["Age"].fillna(df.median()) # median
```

```python
# pie chart
plt.figure(figsize=(10,6))
plt.title("Gender Count")
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
# bar graph 1
plt.figure(figsize=(10,6))
plt.ylabel("Pclass", fontsize=15)
plt.xlabel("Number of people", fontsize=15)
plt.title("Class Comparison", fontsize=15)
df['Pclass'].value_counts().plot(kind='barh', color="#fcca03")
plt.show()
```

```python
# bar graph 2
df.groupby(["Survived", "Pclass"]).size().unstack().plot(kind= "barh", figsize = (8,5))
```

```python
# stacked bargraph
df.groupby(['Sex', ,]).size().unstack().plot(kind='bar', stacked=True, figsize=(9,5))
```

```python
df.groupby(["Survived", "Pclass"]).size().unstack().plot(kind= "barh", figsize = (8,5))
```

```python
# line graph
plt.figure(figsize = (14,7))
plt.style.use("ggplot")
plt.plot(df["Fare"], "g-*",label = "Age")
plt.title("Line plot of fare")
plt.xlabel("n")
plt.ylabel("Fare in dollars")
plt.legend()
plt.show()
```

```python
d = df[["Fare",'Survived']]
sns.boxplot(data = d,x = "Survived",y = "Fare")
```
